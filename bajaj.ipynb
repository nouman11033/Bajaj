```python


import pandas as pd
df = pd.read_json('/content/json_data.json')
df
```

```python

total_medicines = df['consultationData'].apply(lambda x: len(x.get('medicines', []))).sum()


average_medicines = total_medicines / len(df)

print("Total ", total_medicines)
print("Average ", average_medicines)
```

```python

from collections import Counter

all_medicines = []
for consultation in df['consultationData']:
  medicines = consultation.get('medicines', [])
  for medicine in medicines:
    all_medicines.append(medicine['medicineName'])

medicine_counts = Counter(all_medicines)
third_most_common = medicine_counts.most_common(3)[2][0]

print("The 3rd ", third_most_common)
```

```python

mode_gender = df['patientDetails'].apply(lambda x: x.get('gender', '')).mode()[0]

df['patientDetails'] = df['patientDetails'].apply(lambda x: {**x, 'gender': x.get('gender', mode_gender)})

female_count = df['patientDetails'].apply(lambda x: x['gender'] == 'female').sum()
total_count = len(df)
female_percentage = (female_count / total_count) * 100

print("Percentage :", female_percentage)
```

```python
def categorize_age(age):
  if age >= 0 and age <= 12:
    return 'Child'
  elif age >= 13 and age <= 19:
    return 'Teen'
  elif age >= 20 and age <= 59:
    return 'Adult'
  elif age >= 60:
    return 'Senior'
  else:
    return 'Unknown'

df['patientDetails'] = df['patientDetails'].apply(lambda x: {**x, 'ageGroup': categorize_age(x.get('age'))})

adult_count = df['patientDetails'].apply(lambda x: x['ageGroup'] == 'Adult').sum()
print("Count: ", adult_count)
```

```python
active_count = 0
inactive_count = 0

for consultation in df['consultationData']:
  medicines = consultation.get('medicines', [])
  for medicine in medicines:
    if medicine.get('isActive', False):
      active_count += 1
    else:
      inactive_count += 1

total_medicines = active_count + inactive_count

active_percentage = (active_count / total_medicines) * 100
inactive_percentage = (inactive_count / total_medicines) * 100

print("Percentage act {:.2f}%".format(active_percentage))
print("Percentageinact {:.2f}%".format(inactive_percentage))
```

```python
df['num_medicines'] = df['consultationData'].apply(lambda x: len(x.get('medicines', [])))
correlation = df['patientDetails'].apply(lambda x: x.get('age')).corr(df['num_medicines'], method='pearson')
print(correlation)
```

```python

import re

def is_valid_phone(phone):
  pattern = r'^(?:\+91)?[6789]\d{9}$'
  return re.match(pattern, str(phone)) is not None

df['is_valid_phone'] = df['patientDetails'].apply(lambda x: is_valid_phone(x.get('phoneNumber', '')))

print(df[['patientDetails', 'is_valid_phone']])
```

```python
def is_valid_indian_number(phone_number):
  if not isinstance(phone_number, str):
    return False
  if not phone_number.isdigit():
    return False
  if not phone_number.startswith(('6', '7', '8', '9')):
    return False
  return True

df['isValidMobile'] = df['patientDetails'].apply(lambda x: is_valid_indian_number(x.get('phoneNumber', '')))

df.head()
```
